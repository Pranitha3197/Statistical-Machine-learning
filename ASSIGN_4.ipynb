```python
y=df['price_range']
print(y)
print(y.shape)
```

```python
x=df.drop('price_range',axis=1)
print(x)
```

```python
xx=x-x.min()/(x.max()-x.min())
print(xx)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xx_train,xx_test,y_train,y_test=train_test_split(xx,y,test_size=0.2,random_state=30)
print(xx_train)
print(xx_test)
print(y_train)
print(y_test)
```

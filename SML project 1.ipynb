```python
import pandas as pd
import numpy as np
d=pd.read_csv('/content/gym_members_exercise_tracking.csv')
print(d)
```

```python
d['Weight (kg)'] = d['Weight (kg)'].astype(str).str.replace('ND', '0')
d['Height (m)'] = d['Height (m)'].astype(str).str.replace('ND', '0')
d['Max_BPM'] = d['Max_BPM'].astype(str).str.replace('ND', '0')
d['Avg_BPM'] = d['Avg_BPM'].astype(str).str.replace('ND', '0')
d['Resting_BPM'] = d['Resting_BPM'].astype(str).str.replace('ND', '0')
d['Session_Duration (hours)'] = d['Session_Duration (hours)'].astype(str).str.replace('ND', '0')
d['Calories_Burned'] = d['Calories_Burned'].astype(str).str.replace('ND', '0')
d['Workout_Type']=d['Workout_Type'].replace({'Yoga':0,'HIIT':1,'Cardio':2,'Strength':3,'Other':4})
d['Fat_Percentage'] = d['Fat_Percentage'].astype(str).str.replace('ND', '0')
d['Water_Intake (liters)'] = d['Water_Intake (liters)'].astype(str).str.replace('ND', '0')
d['Workout_Frequency (days/week)'] = d['Workout_Frequency (days/week)'].astype(str).str.replace('ND', '0')
d['Experience_Level'] = d['Experience_Level'].astype(str).str.replace('ND', '0')
d['BMI'] = d['BMI'].astype(str).str.replace('ND', '0')
d
```

```python
d.dtypes
```

```python
d['Gender']=d['Gender'].replace({'Male','Female'},{1,0})
d
```

```python
d.fillna(1, inplace=True)
d
```

```python
y=d['BMI']
print(y)
```

```python
x=d.drop('BMI',axis=1)
print(x)
```

```python
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python

accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
y_p=svc.predict(x_test)
y_p
```

```python
accuracy=accuracy_score(y_test,y_p)
accuracy
```

```python
from sklearn.utils import resample
lr_acc=[]
svm_acc=[]
knn_acc=[]
for i in range(0,20):
  x_r,y_r=resample(x,y,n_samples=250)
  x_tr,X_te,y_tr,y_te=train_test_split(x_r,y_r,test_size=0.2,random_state=42)
  model=LogisticRegression()
  model.fit(x_tr,y_tr)
  y_pred=model.predict(X_te)
  lr_acc.append(accuracy_score(y_te,y_pred))
  sv=SVC()
  sv.fit(x_tr,y_tr)
  y_p=sv.predict(X_te)
  svm_acc.append(accuracy_score(y_te,y_p))
  kn=KNeighborsClassifier()
  kn.fit(x_tr,y_tr)
  y_predict=kn.predict(X_te)
  knn_acc.append(accuracy_score(y_te,y_predict))
```

```python

print(lr_acc)
print(svm_acc)
print(knn_acc)
```

```python
import numpy as np
import matplotlib.pyplot as plt
xa=np.arange(0,20)
plt.plot(xa,lr_acc)
plt.plot(xa,svm_acc)
plt.plot(xa,knn_acc)
plt.show()
```

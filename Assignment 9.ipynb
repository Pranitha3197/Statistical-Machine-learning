```python

import pandas as pd
df=pd.read_csv("/content/breast_cancer_survival.csv")
print(df)
```

```python
df['ER status']=df['ER status'].replace({'Negative':0,'Positive':1})
df['PR status']=df['PR status'].replace({'Negative':0,'Positive':1})
df['HER2 status']=df['HER2 status'].replace({'Negative':0,'Positive':1})
df['Gender']=df['Gender'].replace({'MALE':0,'FEMALE':1})
df['Patient_Status']=df['Patient_Status'].replace({'Dead':0,'Alive':1})
df['Histology']=df['Histology'].replace({'Infiltrating Ductal Carcinoma':0,'Infiltrating Lobular Carcinoma':1,'Mucinous Carcinoma':2})
df['Tumour_Stage']=df['Tumour_Stage'].replace({'II':0,'I':1,'III':2})
df['Surgery_type']=df['Surgery_type'].replace({'Lumpectomy':0,'Modified Radical Mastectomy':1,'Simple Mastectomy':2,'Other':3})
```

```python
df
```

```python
df.drop('Date_of_Surgery',inplace=True,axis=1)
df.drop('Date_of_Last_Visit',inplace=True,axis=1)
df
```

```python

y=df['Patient_Status']
y
```

```python
x=df.drop('Patient_Status',axis=1)
x
```

```python
df.head(334)
```

```python
y=df['Patient_Status'].isna()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python

y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred)
print(confusion)
```
